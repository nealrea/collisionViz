```python
import pandas as pd
from geopy.distance import great_circle
collisions = pd.read_csv('../app/collisions.csv')
bikes = pd.read_csv('bike_coll_rate.csv')
```

```python
bike_array = []
for index, row in bikes.iterrows():
    bike_array.append((row['LOCATION'], row['Collision Rate']))
bike_array
```

```python
points = []
coors = []
route_points = [[40.72497, -73.9952], [40.72497, -73.9952], [40.72503, -73.99538], [40.72503, -73.99538], [40.72481, -73.99556], [40.7244, -73.99588], [40.72421, -73.99604], [40.72369, -73.99648], [40.72369, -73.99648], [40.72388, -73.99695], [40.72424, -73.99779], [40.72463, -73.99859], [40.72501, -73.99935], [40.72538, -74.0001], [40.72538, -74.0001]]
for i, ind in enumerate(route_points[::2]):
    danger = 0
    closest = 1000000000
    route_point = (i[0], i[1])
    for point in bike_array:
        data_point = point[0]
        distance = great_circle(route_point, data_point).miles
        new_danger = point[1]
        if (distance < closest) and (new_danger > danger):
            closest = distance
            danger = new_danger
    if ((len(points) < 5) and (route_point not in coors)):
        coors.append(route_point)
        points.append((route_point, danger))
    else:
        for point in range(len(points)):
            if danger > points[point][1]:
                coors.append(route_point)
                points[point] = (route_point, danger)
    points = sorted(points, key=lambda x: x[1])
    print(points)
```
